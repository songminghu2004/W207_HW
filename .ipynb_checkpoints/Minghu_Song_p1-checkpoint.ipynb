(6) A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
### STUDENT START ###
# Gaussian Kernel Function with default std.dev (sigma = 1)
# Return a normalized 2D gauss kernel array for convolutions
# Input, size, defines the dimention (2*size+1) of output kernel array
def gauss_kernel(size, sigma=1):
    """ Returns a normalized 2D gauss kernel array for convolutions """
    size = int(size)              
    #print size, sizey    
    x, y = np.mgrid[-size:size+1, -size:size+1]
    g = (1/(2*np.pi*sigma**2))* np.exp(-(x**2 + y**2)/(2 * sigma**2))
    return g / g.sum()  # normalization by the sum of array

# Gaussian Blur Function to apply to each pixel in the image.
# Return a transformed image data with the same size as the input data
# Input: image array data and a gussian keneral array  
def gauss_blur(imagearray, kernel_weight):        
    blurData = np.zeros(imagearray.shape)
    # For each row or image within the original data
    for i in range(imagearray.shape[0]):
        # Reshape to 28x28 matrix
        mat = imagearray[i].reshape(28,28)
        blurMat = np.zeros((28, 28))
        for j in range(1, 27):
            for k in range(1, 27):
                for l in range(-1, 2):
                    for m in range(-1, 2):
                        # Add weighted value to blurred matrix
                        blurMat[j][k] += mat[j+l][k+m] * kernel_weight[l+1][m+1]
        # Reshape the blurred matrix back to the original dimension
        blurData[i] = blurMat.reshape(1, 784)
    return blurData    

# KNN Classification wrapper for given trainX, TrainY, testX, testY 
def clf_wrapper(trainX, trainY, testX, testY):
        # Setup classifier and check accuracy for each case
        clf = KNeighborsClassifier(n_neighbors=1)
        clf.fit(trainX, trainY) 
        preds = clf.predict(testX)
        correct, total = 0, 0
        for pred, label in zip(preds, testY):
            if pred == label: correct += 1
            total += 1
        print('total: %3d  correct: %3d  accuracy: %4.3f' %(total, correct, 1.0*correct/total)  ) 
        
def P6():
    # Return a normalized 3*3 gauss kernel ndarray as weights for convolutions
    kernel = gauss_kernel(1)
    
    # Data preprocessing throught the Gussian Blurring filter 
    blurred_mini_train_data = gauss_blur(mini_train_data, kernel)
    blurred_dev_data = gauss_blur(dev_data, kernel) 

    # Visulize outcomes of the Gussian blurring filter before modelling
    # Print out some examples of raw digits before Gaussian blurring for comparision
    print('Examples of digits before Gussian blurring:')
    plt.figure(figsize = (15,6))
    for subplot_index in range(15):
        mat1 = dev_data[subplot_index,].reshape(28,28)
        plt.subplot(1, 15, subplot_index+1)
        plt.imshow(mat1, cmap = 'Greys')
        plt.axis('off')
    plt.show()
    # Print out some examples of the same digits after the Gaussian blurring process
    print('Examples of digits after Gussian blurring:')
    plt.figure(figsize = (15,6))
    index=0
    for subplot_index in range(15):
        mat2 = blurred_dev_data[subplot_index,].reshape(28,28)
        plt.subplot(1, 15, subplot_index+1)
        plt.imshow(mat2, cmap = 'Greys')
        plt.axis('off')
    plt.show()
    
    # Now let's print the model performanc before any blurring process 
    print('Modelling with preprocessing training and dev data:')
    clf_wrapper(mini_train_data, mini_train_labels, dev_data, dev_labels)
    print('\n')
    
    # Now let's print the model performanc after certain blurring process 
    print( 'Modelling with preprocessing training data but not dev data:')
    clf_wrapper(blurred_mini_train_data, mini_train_labels, dev_data, dev_labels)
    print('Modelling with preprocessing dev data but not training data:' )
    clf_wrapper(mini_train_data, mini_train_labels, blurred_dev_data, dev_labels)
    print('Modelling with preprocessing both training and dev data:')
    clf_wrapper(blurred_mini_train_data, mini_train_labels, blurred_dev_data, dev_labels)

    ### STUDENT END ###

P6()
```

ANSWER: In this exercise, we set the default sigma (std) = 1. As shown in picturs, Blurring filters (eliminates) high-frequency noise, reducing details which will makes the model more generalizable and is less prone to overfitting. 

Our results show that, compared with the unprocessing one, the KNN accuracy can be improved (or at least slightly improved) when both train and dev data are processed with the gaussian blurring filter. Even when only the training set is preprocessed and the dev data remains intact, we still obtain a better model. However, when we only process the dev data instead of the train one, the prediction performance decreases.  In that case, we maintain the details in the training set while remove the details in the dev set.  This can lead to the loss of certain information within the dev set before the modeling fitting. 

(7) Fit a Naive Bayes classifier and report accuracy on the dev data. Remember that Naive Bayes estimates P(feature|label). While sklearn can handle real-valued features, let's start by mapping the pixel values to either 0 or 1. You can do this as a preprocessing step, or with the binarize argument. With binary-valued features, you can use BernoulliNB. Next try mapping the pixel values to 0, 1, or 2, representing white, grey, or black. This mapping requires MultinomialNB. Does the multi-class version improve the results? Why or why not?

```python
# Function to pixel binarization or trinarization for given parameter n 
def pixel_mapping(image_array, n):
    #temp = np.copy(image_array)
    """
    Binarization (n=2): Map the pixel values to 0, 1.
    """
    if n == 2:
        #temp = np.copy(image_array)
        image_array[image_array <= 1.0/2 ] = 0
        image_array[image_array > 1.0/2 ] = 1
        return image_array
    
    """
    Trinarization (n=3): Map the pixel values to 0, 1, 2.
    """
    if n == 3:
        #temp = np.copy(image_array)
        image_array[image_array <= 1.0/3] = 0  # white 
        image_array[(image_array > 1.0/3) & (image_array <= 2.0/3) ] = 1 # grey
        #image_array[np.logical_and(image_array > 1.0/3, image_array <= 2.0/3)] = 1 # use logic function
        image_array[image_array > 2.0/3] = 2 # black
        return image_array


def P7():
    # Create a BernoulliNB instance and map the pixel values to 1 or 1 with the binarize parameter of 0.5 
    binomialnb_model = BernoulliNB(binarize=0.5)
    binomialnb_model.fit(train_data, train_labels)
    # Obtain the BernoulliNB model accuracy
    print('Accuracy of the BernoulliNB model on the traing data: %.3f' %(binomialnb_model.score(train_data, train_labels)) )
    print('Accuracy of the BernoulliNB model on the dev data: %.3f' %(binomialnb_model.score(dev_data, dev_labels) ))
    print('\n')
    
    # Alternatively, call the pixel_mapping function for the binarization
    # bin_training_data = [pixel_mapping(image, 2) for image in train_data]
    # bin_dev_data = [pixel_mapping(image, 2) for image in dev_data]
    # clf = BernoulliNB()
    # clf.fit(bin_training_data, train_labels)
    # print("bernoulli: ", clf.score(bin_dev_data, dev_labels) )
    
    #  Call the pixel_mapping function for the trinarization
    tri_training_data = [pixel_mapping(image, 3) for image in train_data]
    tri_dev_data = [pixel_mapping(image, 3) for image in dev_data]
    multinomialnb_model = MultinomialNB()
    multinomialnb_model.fit(tri_training_data, train_labels)
    print('Accuracy of the Multinomial model on the train data: %.3f' %(multinomialnb_model.score(tri_training_data, train_labels) ) )
    print('Accuracy of the Multinomial model on the dev data: %.3f' %(multinomialnb_model.score(tri_dev_data, dev_labels) ) )
    
P7()
```

ANSWER: As shown above, the multi-class version doesn't improve the model accuracy.  Instead, the BernoulliNB model gives a higher accuracy on both dev and traing data than the Multinomia model.  In the BernoulliNB method, features are independent booleans, whereas the multinomial model keeps track of numtiple occurences. In this case, BernoulliNB with bollean features would be more  precise to describe digits than the multinomial method.  In addition, those nonoccurencing features are ofen taken into the consideration and explicitly penalized in the BernoulliNB model. On the other hand, the multinomial variant simply ignores those non-occurring features, which often result in certain information loss.  Therefore, the BernoulliNB model  outperforms the multinomial model in this case.

ANSWER: The best value obtained from GridSearchCV for the mini_train_data is 0.01.   When alpha is equal to 0, the obtained mean accuracy is abot 0.092.  Based on the above results,  it seems that a small alpha ( e.g. 0.001 or 0.01) would  produce the best mean CV score.  When alpha = 0, its mean accuracy is quite low (~0.092).
```python
def P9():
### STUDENT START ###
    # Train a GaussianNB classifier
    gnb = GaussianNB()
    gnb.fit(mini_train_data, mini_train_labels)
    print('Accuracy of original GaussianNB model is : %.3f' %(gnb.score(dev_data, dev_labels)) )

    # Randomly select a digit for diagonosis 
    # Plot frequency histogram of the random image
    # Then add white noise to the train and dev data for comparision 
    rand_image = X[np.random.choice(X.shape[0])]
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    fig = plt.hist(rand_image,100)
    plt.title('Histogram of a randomly selected digit')
    plt.subplot(1, 2, 2)
    fig = plt.hist(rand_image + np.random.normal(0, 0.2, rand_image.shape), 100)
    plt.title('Histogram of the random digit with added white noise')
    
    # Examine the best parameter (sigma) with maximized accuracy
    sigmas = np.arange(0.05, 1.0, 0.05)
    accuracies = np.zeros(sigmas.shape)
    for i in range(sigmas.shape[0]):
        new_train_data = mini_train_data + np.random.normal(0,sigmas[i], mini_train_data.shape)
        new_dev_data = dev_data + np.random.normal(0,sigmas[i], dev_data.shape)
        gnb.fit(new_train_data, mini_train_labels)
        accuracies[i] = gnb.score(new_dev_data, dev_labels)
    best_sigma=sigmas[accuracies.argmax()]
    print('Best approximate sigma: %.2f' %(best_sigma) )
    
    # Tranformation:  Add white noises to train and dev data before fitting
    new_train_data = mini_train_data + np.random.normal(0,best_sigma,mini_train_data.shape)
    new_dev_data = dev_data + np.random.normal(0,best_sigma,dev_data.shape)
    gnb.fit(new_train_data, mini_train_labels)
    print('Accuracy of new GaussianNB model after adding white noises is: %.3f' %(gnb.score(new_dev_data, dev_labels)) )
### STUDENT END ###

gnb = P9()
```

ANSWER:  The Gaussian NB model assumes that features follow a normal distribution.  However, this is not correct in our current case as shown in the left figure. Therefore, we will obtain a poor performance (0.635) because features don't follow the normal distribution. In this case, we should use transformation or different methods to convert the training and dev data in normal distribution. For example, we can add gaussian noise with zero mean and a standard deviation that maximizes the accuracy on the dev set. After the data tranforming, we obtain a better model accuracy (~0.804)

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counter.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets, correct, total):
    
### STUDENT START ###
    # Create a Binomial NB instance and fit it with train data 
    bnb = BernoulliNB(alpha=0.0001, binarize=0.5)
    bnb.fit(train_data, train_labels)
    
    # Predict labels and probabilities of dev set
    dev_predlabels = bnb.predict(dev_data)
    dev_predprobs  = bnb.predict_proba(dev_data)
    
    # Loop through each posterior bucket and then count variables of correct and total for each bucket
    prev_bucket = 0
    for bucket_index, bucket in enumerate(buckets):
        for i, value in enumerate(dev_predprobs):
            if value.max() <= bucket and value.max() > prev_bucket:
                total[bucket_index] += 1
                if dev_predlabels[i] == dev_labels[i]:
                    correct[bucket_index] += 1
        prev_bucket = bucket
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): accuracy = correct[i] / total[i]
    print('p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy) )
```

ANSWER: Although the accuracy increases when the posterior probability of the predict class increases, our BernoulliNB classifier is a weakly calibrated classifier in this case because the accuracies is only 0.442 and 0.483 even when the posterior probability of predicted class is 0.9 and 0.999, respectively.

```python
### STUDENT START ###
from scipy import ndimage
from sklearn import preprocessing
# Suppress Python3.5 RuntimeWarning message
import warnings
warnings.filterwarnings("ignore")

# Function to count pixel values in each row/column and return 56 features 
def pixel_sum(data):
    # Create an array with the right shape to hold new feature data
    newData = np.empty((data.shape[0], 28*2), float)

    # Will transforms indivisual features by scaling them to the default range (0,1)
    scaler = preprocessing.MinMaxScaler()
    
    # Go through each digit, find out the list of row sums and column sums, and add them as features
    for i in range(data.shape[0]):
        imgf = ndimage.gaussian_filter(data[i,].reshape(28, 28), 0.5)
        col_sum = np.sum(imgf, axis=0)
        row_sum = np.sum(imgf, axis=1)

        # Transform counts to fall between 0 and 1 as we're using Bernoulli NB
        col_sum = scaler.fit_transform(col_sum)
        row_sum = scaler.fit_transform(row_sum)
       
        # Append two vectors, col_sum and row_sum, into ith row of newdata 
        #newData[i,] = np.append(np.append(data[i], col_sum), row_sum)
        newData[i] =np.append(col_sum, row_sum)   
    return newData

# Function to count number of enclosed regions in digit
# here I use the ndimage.label function to recognize the empty region 
# through the standard two-pass binary connected-component labeling algorithm
# In this case, this function is used to identify subsets of connected components
# (or objects) that are uniquely labeled with pixel values less than a cutoff.
# Since derived object numbers will also include the boundary regions where have the 
# low pixel values as well, we have to substract the number by one to get the number of holes. 
def hole_sum(data):
    threshold = 0.5
    # Create an array with the right shape
    newData = np.empty((data.shape[0], 1), int)
 
    scaler = preprocessing.MinMaxScaler()

    # Go through each digit image, find out numer of empty regions within the digit image
    # return the number of holes fully enclosed by the digit where its pixel values are 
    # usually larger than the cutoff
    for i in range(data.shape[0]):
        imgf = data[i,].reshape(28, 28)
        labeled, nr_objects = ndimage.label(imgf < threshold) 
        #print('Number of objects is %d' %(nr_objects-1) )
        newData[i,]= nr_objects-1
    return newData

def P12():
    # Print the performance of original BernoulliN model on the dev set
    clf = BernoulliNB(binarize=0.5, alpha=0.01)
    clf = clf.fit(mini_train_data, mini_train_labels)
    print('%s: Accuracy = %.3f' %("Performance of original BernoulliN model on the dev set", clf.score(dev_data, dev_labels)) )

    # Generate 56 new features: sum the pixel values in each row and each column
    # Print the performance of new model with added features
    mini_train_data_pixelsum = pixel_sum(mini_train_data)
    mini_train_data_2 = np.append(mini_train_data, mini_train_data_pixelsum,  axis=1 )   
    dev_data_pixelsum = pixel_sum(dev_data)
    dev_data_2 = np.append(dev_data, dev_data_pixelsum,axis =1 )
    clf = clf.fit(mini_train_data_2, mini_train_labels)
    print('%s: Accuracy = %.3f' %("Performance of the models with features counting pixel values in each row/column.", clf.score(dev_data_2, dev_labels)) )
    
    # Count the number of empty regions that are enclosed by the digit
    # Print the performance of new model with two features sets:
    # features: sum the pixel values in each row and each column as well as the number of holes
    mini_train_data_holesum = hole_sum(mini_train_data)
    mini_train_data_3 = np.append(mini_train_data_2, mini_train_data_holesum,  axis=1)
    dev_data_holesum = hole_sum(dev_data)
    dev_data_3 = np.append(dev_data_2, dev_data_holesum, axis =1 )
    clf = clf.fit(mini_train_data_3, mini_train_labels)
    print('%s: Accuracy = %.3f' %("Performance of the model with features counting pixel values and the hole number", clf.score(dev_data_3, dev_labels)) )

    # Print 40 examples of digits and counts of enclosed regions
    plt.figure(figsize = (15,6))
    hole_example = hole_sum(dev_data[:40,])
    subplot_index =0
    # Print examples of digits and identified empty regions
    for i in range(40):
        hole_num = hole_example[i][0]
        subplot_index += 1
        image = dev_data[i,].reshape(28,28)
        plt.subplot(4, 10, subplot_index)
        plt.imshow(image)
        plt.axis('off')
        title = '#Hole: %d' %(hole_num)
        plt.title(title)
### STUDENT END ###

P12()
```
