```python
# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

# Import a bunch of libraries.
import time
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from sklearn.pipeline import Pipeline
from sklearn.datasets import fetch_mldata
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report

# Set the randomizer seed so results are the same each time.
np.random.seed(0)
```

```python
# Load the digit data either from mldata.org, or once downloaded to data_home, from disk. The data is about 53MB so this cell
# should take a while the first time your run it.
mnist = fetch_mldata('MNIST original', data_home='~/datasets/mnist')
X, Y = mnist.data, mnist.target

# Rescale grayscale values to [0,1].
X = X / 255.0

# Shuffle the input: create a random permutation of the integers between 0 and the number of data points and apply this
# permutation to X and Y.
# NOTE: Each time you run this cell, you'll re-shuffle the data, resulting in a different ordering.
shuffle = np.random.permutation(np.arange(X.shape[0]))
X, Y = X[shuffle], Y[shuffle]

print('data shape: ', X.shape )
print('label shape:', Y.shape )
print(Y[:200])

# Set some variables to hold test, dev, and training data.
test_data, test_labels = X[61000:], Y[61000:]
dev_data, dev_labels = X[60000:61000], Y[60000:61000]
train_data, train_labels = X[:60000], Y[:60000]
mini_train_data, mini_train_labels = X[:1000], Y[:1000]
```

```python
### STUDENT START ###
from sklearn import preprocessing
# Suppress Python3.5 RuntimeWarning 
import warnings
warnings.filterwarnings("ignore")

# Function to count pixel values in each row/column and return 56 features 
def pixel_sum(data):
    # Create an array with the right shape to hold new feature data
    newData = np.empty((data.shape[0], 28*2), float)

    # Will transforms indivisual features by scaling them to the default range (0,1)
    scaler = preprocessing.MinMaxScaler()
    
    # Go through each digit, find out the list of row sums and column sums, and add them as features
    for i in range(data.shape[0]):
        imgf = ndimage.gaussian_filter(data[i,].reshape(28, 28), 0.5)
        col_sum = np.sum(imgf, axis=0)
        row_sum = np.sum(imgf, axis=1)

        # Transform counts to fall between 0 and 1 as we're using Bernoulli NB
        col_sum = scaler.fit_transform(col_sum)
        row_sum = scaler.fit_transform(row_sum)
       
        # Append two vectors, col_sum and row_sum, into ith row of newdata 
        #newData[i,] = np.append(np.append(data[i], col_sum), row_sum)
        newData[i] =np.append(col_sum, row_sum)   
    return newData

import scipy
from scipy import ndimage

def hole_sum(data):
    threshold = 0.5
    # Create an array with the right shape
    newData = np.empty((data.shape[0], 1), int)
 
    scaler = preprocessing.MinMaxScaler()

    # Go through each digit, find out numer of empty regions within the digit image
    for i in range(data.shape[0]):
        imgf = data[i,].reshape(28, 28)
        labeled, nr_objects = ndimage.label(imgf < threshold) 
        #print('Number of objects is %d' %(nr_objects-1) )
        newData[i,]= nr_objects-1
    return newData

def P12():
    # Print the performance of original BernoulliN model on the dev set
    clf = BernoulliNB(binarize=0.5, alpha=0.01)
    clf = clf.fit(mini_train_data, mini_train_labels)
    print('%s: Accuracy = %.3f' %("Performance of original BernoulliN model on the dev set", clf.score(dev_data, dev_labels)) )

    # Generate 56 new features: sum the pixel values in each row and each column
    # Print the performance of new model with added features
    mini_train_data_pixelsum = pixel_sum(mini_train_data)
    mini_train_data_2 = np.append(mini_train_data, mini_train_data_pixelsum,  axis=1 )   
    dev_data_pixelsum = pixel_sum(dev_data)
    dev_data_2 = np.append(dev_data, dev_data_pixelsum,axis =1 )
    clf = clf.fit(mini_train_data_2, mini_train_labels)
    print('%s: Accuracy = %.3f' %("Performance of the models with extra features that count the pixel values in each row and each column.", clf.score(dev_data_2, dev_labels)) )
    
    # Count the number of empty regions in the digit image
    # Print the performance of new model with two features sets
    mini_train_data_holesum = hole_sum(mini_train_data)
    mini_train_data_3 = np.append(mini_train_data_2, mini_train_data_holesum,  axis=1)
    dev_data_holesum = hole_sum(dev_data)
    dev_data_3 = np.append(dev_data_2, dev_data_holesum, axis =1 )
    clf = clf.fit(mini_train_data_3, mini_train_labels)
    print('%s: Accuracy = %.3f' %("Performance of the model with features counting pixel values in each row/col and the number of empty regions", clf.score(dev_data_3, dev_labels)) )

    plt.figure(figsize = (10,6))
    hole_example = hole_sum(dev_data[:30,])
    subplot_index =0
    # Show examples of digits and identify empty regions
    for i in range(30):
        print(hole_example[i][0])
        hole_num = hole_example[i][0]
        subplot_index += 1
        image = dev_data[i,].reshape(28,28)
        plt.subplot(5, 6, subplot_index)
        plt.imshow(image)
        plt.axis('off')
        title = '#Hole: %d' %(hole_num)
        plt.title(title)

### STUDENT END ###

P12()
```
